```python
x = df.drop(columns=["price"], axis=1)
y = df["price"]
```

```python
x
```

```python
x_one_encode = pd.get_dummies(x,columns =['model','transmission','fuelType'], drop_first = True)
```

```python
x_one_encode = x_one_encode.astype(int)
```

```python
x_one_encode
```

```python
from sklearn.preprocessing import LabelEncoder

columns = ['model', 'transmission', 'fuelType']

xlabel = x.copy()        # correct copy
label_encoders = {}     # dictionary to store encoders

for col in columns:
    le = LabelEncoder()
    xlabel[col] = le.fit_transform(xlabel[col].astype(str))
    label_encoders[col] = le
```

```python
 xlabel
```

```python
df['model'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
```

```python
scaler = StandardScaler()
x_one_encode[numerical_cols] = scaler.fit_transform(x_one_encode[numerical_cols])
```

```python
x_one_encode
```

```python
xlabel.columns
```

```python
xlabel[['model', 'year', 'transmission', 'mileage', 'engineSize']] = (
    scaler.fit_transform(
        xlabel[['model', 'year', 'transmission', 'mileage', 'engineSize']]
    )
)

```

```python
xlabel
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

```

```python
 x_train, x_test, y_train, y_test = train_test_split(x_one_encode, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# 1 Train the model
lr = LinearRegression()
lr.fit(x_train, y_train)

# 2 Make predictions
y_pred = lr.predict(x_test)

# 3 Now this will work
y_pred
```

```python
x_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_pred)
r2
```

```python
n = x_test.shape[0]   # number of samples
p = x_test.shape[1]   # number of features

adjusted_r2 = 1 - ((1 - r2) * (n - 1)) / (n - p - 1)

print("Adjusted R2 score:", adjusted_r2)
```

```python
# 2nd model tren test
x_train, x_test, y_train, y_test = train_test_split(xlabel, y, test_size=0.33, random_state=42)
```

```python
model2 = LinearRegression()
model2.fit(x_train, y_train)
```
